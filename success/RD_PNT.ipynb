```julia
@model rd_pnt begin
    df = CSV.read("road_surface.csv", DataFrame)
    @parameter begin
        kmph2mps = 0.27778
        I_wheelbase=2.635
        vel_car_kmph=60
        end_of_road_surface=43  ###
    end
    @xmlmodel("METI_Vertical_Vibration.drawio", "RD_PNT")
end
```
